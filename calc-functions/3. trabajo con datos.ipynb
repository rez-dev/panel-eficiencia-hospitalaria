```python
import pysfa
import pandas as pd
import numpy as np

# cargar datos consolidados
df = pd.read_csv('df_consolidado_final_v2.csv', sep=',')

# pasar valores con -- a 0 en todas las columnas
df = df.replace({'--': 0})

# para las columnas Consulta, GRDxEgreso, Bienes y servicios, Remuneraciones, Dias Cama Disponibles, Consultas Urgencias, Examenes y Quirofanos transformar missing en 0
df['Consultas'] = df['Consultas'].fillna(0)
df['GRDxEgreso'] = df['GRDxEgreso'].fillna(0)
df['Bienes y servicios'] = df['Bienes y servicios'].fillna(0)
df['Remuneraciones'] = df['Remuneraciones'].fillna(0)
df['Dias Cama Disponibles'] = df['Dias Cama Disponibles'].fillna(0)
df['Consultas Urgencias'] = df['Consultas Urgencias'].fillna(0)
df['Examenes'] = df['Examenes'].fillna(0)
df['Quirofanos'] = df['Quirofanos'].fillna(0)

# drop filas con 0 en alguna columna
df = df.dropna()
```

```python
# pasar columna Consultas a float
df['Consultas'] = df['Consultas'].astype(int)
df['GRDxEgreso'] = df['GRDxEgreso'].astype(int)
df['Bienes y servicios'] = df['Bienes y servicios'].astype(int)
df['Remuneraciones'] = df['Remuneraciones'].astype(int)
df['Dias Cama Disponibles'] = df['Dias Cama Disponibles'].astype(int)
df['Consultas Urgencias'] = df['Consultas Urgencias'].astype(int)
df['Examenes'] = df['Examenes'].astype(int)
df['Quirofanos'] = df['Quirofanos'].astype(int)
```

```python
# filtrar datos solo con año 2014
df = df[df['Año'] == 2014]

# obtener una muestra de 10 hospitales
df = df.sample(n=10, random_state=1)
```

```python
import numpy as np
# 2. Definir log-variables

# Asegura que las columnas sean numéricas
df['Remuneraciones'] = pd.to_numeric(df['Remuneraciones'], errors='coerce')
df['Bienes y servicios'] = pd.to_numeric(df['Bienes y servicios'], errors='coerce')
df['Dias Cama Disponibles'] = pd.to_numeric(df['Dias Cama Disponibles'], errors='coerce')
df['Consultas'] = pd.to_numeric(df['Consultas'], errors='coerce')

x = np.log(df[['Remuneraciones', 'Bienes y servicios', 'Dias Cama Disponibles']]).to_numpy()
y = np.log(df['Consultas']).to_numpy()
```

```python
x
```

```python
cols = ["Remuneraciones", "Bienes y servicios", "Dias Cama Disponibles", "Consultas"]

# identifica filas con ±inf o números no finitos
mask_bad = ~np.isfinite(df[cols].to_numpy())
rows_bad = np.where(mask_bad.any(axis=1))[0]

print("Filas problemáticas:", rows_bad)        # índices con inf
print(df.iloc[rows_bad][cols])
```

```python
df[cols] = df[cols].replace(0, np.nan)          # evita -inf
df = df.replace([np.inf, -np.inf], np.nan)      # purga infinitos

# Imputación rápida o eliminación:
df = df.dropna(subset=cols)                     # opción 1: descartar
# o bien:
# df[cols] = df[cols].fillna(df[cols].median())   # opción 2: imputar
```

```python
from pysfa import SFA

# 3. Crear y optimizar SFA
sfa = SFA.SFA(y, x, fun=SFA.FUN_PROD, method=SFA.TE_teJ)
sfa.optimize()
# 4. Añadir eficiencia al DataFrame
df['Eff_SFA'] = sfa.get_technical_efficiency()
# 5. Guardar para el dashboard
df[['hospital_name', 'Eff_SFA']].to_csv("sfa_scores.csv", index=False)

# print estimates
print(sfa.get_beta())
print(sfa.get_residuals())

# print estimated parameters
print(sfa.get_lambda())
print(sfa.get_sigma2())
print(sfa.get_sigmau2())
print(sfa.get_sigmav2())

# print statistics
print(sfa.get_pvalue())
print(sfa.get_tvalue())
print(sfa.get_std_err())

# OR print summary
print(sfa.summary())

# print TE
print(sfa.get_technical_efficiency())
```
